## **En yakın ada**


Bu serüvenimizde sizler okyanusun ortasında ve suyun içindesiniz. Henüz tam olarak yüzmeyi bilmiyorsunuz (can yeleğiniz var). Su, yemek ve kurtuluş adında üç adaya yüzmeyi öğrenip sırasıyla ulaşmak zorundasınız. En yakın adanın adı ***SU***.
<br><br>
Su'ya ulaşmak için;
 - fonksiyonun hatasız bir makine olduğunu kavramalı ve fonksiyonlara dair kavramlara hakim olmalısınız,
 - çizili fonksiyonlarda argümanları manipüle etmenin yaratacağı sonuçları okuyabiliyor olmalısınız,
 - argümanlara tepkisi bilinen fonksiyonları kabaca çizebiliyor olmalısınız,
 - formülü bilinen fonksiyonları doğrulukla çizebiliyor olmalısınız,
 - birinci ve ikinci türev bilgisinin ne anlama (işlev ve yorum) geldiğini biliyor olmalısınız,
 - fonksiyon grafiklerinde birinci ve ikinci türev bilgisinin hangi bölgelerde var olduğunu, var ise pozitif ya da negatif olduğunu sezebiliyor olmalısınız.
<br><br>
Su'ya ulaştığımızda "optimizasyon" ve "arama" kavramları ile tanışmaya hazır olacaksınız.
<br><br>
Bu iki kavram doğru karar vermede iki en güçlü ve iki en kritik silahınız olacak.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = [8,4.5]
plt.rcParams['figure.dpi'] = 100
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False

#  Number of DMUs
N = 10**3

#  Number of inputs
m = 2

#  Number of outputs
s = 1

#    creating a product operator function
from functools import reduce
import operator
def prod(iterable):
    return reduce(operator.mul, iterable, 1)

np.random.seed(101)

#  randomized input factors in production
# p = 5 + abs( np.random.randn( m ) )
# p = { j:[p[i]/p[i].sum() for i in range(m)] for j in range(N) }
# p = np.array( [p[j] for j in range(N)] )
p = np.array( [.5,.5,] )

#  random input data
input = np.random.uniform(1,100, size = (N,m))

#    calculating output via a four-input-Cobb-Douglass Production Function
output = np.array(
[ prod(input[j][i]**p[i] for i in range(m)) for j in range(N) ]
)

#  A dataframe of DEA data
df = pd.DataFrame( index = range(N), columns = ['Input1','Input2','Output1'] )

df.Input1 = pd.Series( [input[j][0] for j in range(N)] )
df.Input2 = pd.Series( [input[j][1] for j in range(N)] )

df.Output1 = pd.Series( output )

df

# sns.set()
sns.scatterplot( x = df.Input1, y = df.Output1, alpha = .5, c = df.Input2, cmap = 'coolwarm' )

plt.show()
```
