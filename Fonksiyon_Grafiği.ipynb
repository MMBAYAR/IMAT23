# Fonksiyon Grafiği Çizimi

## **1. Fonksiyon Grafiğini Sağa ya da Sola Kaydırma**

$c \geq 0$ olsun ve $f(x)$'in grafiğinin elimizde olduğunu varsayalım ($f(x)$ biliniyorsa grafiğini çizebiliriz...). $f(x)$'in grafiğini c birim sağa ya da sola kaydırmak için sırasıyla aşağıdaki iki fonksiyonun grafiği çizilir.

* $f(x)$'in grafiğini c birim $\Big\{{\begin{matrix}sağa\ kaydırmak\ f(x-c)\ 'nin\\sola\ kaydırmak\ f(x+c)\ 'nin\end{matrix}}$ grafiğini verir.

<br>**Ör1.**
<br>$f(x)=x^2+x$ için fonksiyonu c birim sağa kaydırmak için 'x' yerine 'x-c' yazılmalıdır:
<br>$f(x)=x^2+x\ ⇒\ {\begin{array}{lll}f(x-c)&=&(x-c)^2+(x-c) 
\\ &=&(x^2-2cx+c^2)+(x-c)
\\ &=&x^2+(1-2c)x+c^2-c
\end{array}}$
<br>$c=1$ için;
<br>\begin{array}{lll}
f(x-1)&=&(x-1)^2+(x-1)\\&=&x^2-x
\end{array}
<br>olur.
<br>
<br>**Ör2.**
<br>$f(x)=x^2-2x+1$'in grafiğini çiziniz ve elde ettiğiniz grafiği 2br. sola kaydırınız.


## **2. Fonksiyon Grafiğini Yukarı ya da Aşağı Kaydırma**

$c \geq 0$ olsun ve $f(x)$'in grafiğinin elimizde olduğunu varsayalım ($f(x)$ biliniyorsa grafiğini çizebiliriz...). $f(x)$'in grafiğini c birim yukarı ya da aşağı kaydırmak için sırasıyla aşağıdaki iki fonksiyonun grafiği çizilir.
* $f(x)$'in grafiğini c birim $\Big\{{\begin{matrix}yukarı\ kaydırmak\ f(x)+c\ 'nin\\aşağı\ kaydırmak\ f(x)-c\ 'nin\end{matrix}}$ grafiğini verir.

<br>**Ör3.**
<br>$f(x)=1+\sqrt{x-1}$'in grafiğini çiziniz ve elde ettiğiniz grafiği 1 br. yukarı kaydırınız.

<br>**Ör4.**
<br>$y=x^2+4x+1$'in grafiğini çiziniz.
<br>Not: verilen fonksiyon bir tam kare ifadesine benzemektedir ve bu ifadeyi tam kare yapan işlem fonksiyon grafiğini yukarı ya da aşağı kaydırma operasyonu ile benzeşmektedir.
<br><br>\begin{array}{lll}y=x^2+4x+1&=&(x^2+4x+4)-3 
\\ &=&(x+2)^2-3
\end{array}
<br>öyleyse bu fonksiyonu çizmek $y=(x+2)^2$'nin grafiğini 3br. aşağı kaydırarak çizmek ile denktir. 


## **3. Fonksiyon Grafiğini Genişletmek ya da Daraltmak**

$c \geq 0$ olsun ve $f(x)$'in grafiğinin elimizde olduğunu varsayalım ($f(x)$ biliniyorsa grafiğini çizebiliriz...). $f(x)$'in grafiğini x ekseni üzerinde c faktörü ile genişletmek ya da daraltmak için sırasıyla aşağıdaki iki fonksiyonun grafiği çizilir.
* $f(x)$'in grafiğini x ekseni üzerinde c faktörü ile $\Big\{{\begin{matrix}genişletmek\ f(x/c)\ 'nin\\daraltmak\ f(cx)\ 'in\end{matrix}}$ grafiğini verir.

<br>**Ör5.**
<br>${{1}\over{2x-1}}$'in grafiğini çiziniz ve elde ettiğiniz grafiği x ekseni üzerinde 2 faktörü ile daraltınız.

**NOT:** Yeni fonksiyonun $x_0$'daki değeri, $f(x)$'in $x_0 / c$'deki değerine eşittir.

$f(x)$'in grafiğini y ekseni üzerinde c faktörü ile genişletmek ya da daraltmak için sırasıyla aşağıdaki iki fonksiyonun grafiği çizilir.
* $f(x)$'in grafiğini y ekseni üzerinde c faktörü ile $\Big\{{\begin{matrix}genişletmek\ cf(x)\ 'in\\daraltmak\ f(x)/c\ 'nin\end{matrix}}$ grafiğini verir.

<br>**Ör6.**
<br>${{1}\over{2x-1}}$'in grafiğini çiziniz ve elde ettiğiniz grafiği y ekseni üzerinde 2 faktörü ile daraltınız.

## **4. Fonksiyon Grafiğini Yansıtmak**

$f(x)$'in garfiğini y üzerinden yansıtmak için (y eksenine göre simetriğini çizmek için) fonksiyonun her $({x,f(x)})$ noktası $({-x,f(x)})$ taşınarak $f(x)$'in grafiği $f(-x)$'in grafiğine dönüştürülür.

<br><br>Benzer biçimde $f(x)$'in grafiğini x ekseni üzerinden yansıtmak için fonksiyonun her $({x,f(x)})$ noktası $({x,-f(x)})$ taşınarak $f(x)$'in grafiği $-f(x)$'in grafiğine dönüştürülür.

<br><br>Ve son olarak $f(x)$'in grafiğini orijin noktası üzerinden yansıtmak için fonksiyonun her $({x,f(x)})$ noktası $({-x,-f(x)})$ taşınarak $f(x)$'in grafiği $-f(-x)$'in grafiğine dönüştürülür.

<br><br>
* $f(x)$'in grafiğini $\Bigg\{
{\begin{array}{l}y\ ekseni\ üzerinden\ yansıtmak\ için\ f(-x)\ 'in\\
x\ ekseni\ üzerinden\ yansıtmak\ için\ -f(x)\ 'in\\
orijin\ noktası\ üzerinden\ yansıtmak\ için\ -f(-x)\ 'in
\end{array}}$ grafiğini verir.

<br><br>
* $f(-x) = f(x)$ ise $f(x)$ için çifttir,
* $f-(x) = -f(x)$ ise $f(x)$ için tektir denir.

<br>**Ör7.**
<br>${{1}\over{2x-1}}$'in grafiğini çiziniz ve elde ettiğiniz grafiği x ekseni üzerinde 2 faktörü ile daraltınız.
