<br><br>***Tanım kümesi (D)***, fonksiyon argümanlarının alabildiği tüm değerleri içeren kümedir.<br><br>
- **Değer Kümesi (S)** Yukarıda grafiği verilmiş olan $f(x)$ için ***değer aralığı*** $[-1.072,\ 2.924]$'tür.
<br><br>***Değer kümesi***, tanım kümesindeki tüm argüman değerleri için fonksiyonun alabildiği değerlerin kümesidir ( $[min(f(x)),\ maks(f(x))]$ ).<br><br>
- **Uç Nokta :** Yukarıda grafiği verilmiş olan $f(x)$ için uç noktalar soldan sağa doğru $(-0.3,\ -0.958)$ ve $(5,\ 2.412)$ noktalarıdır.
<br><br>Tanım kümesindeki en küçük ve en büyük değerler ve bu değerlere karşılık gelen fonksiyon değerleri ile koordinatları tanımlanan noktalardır.
<br><br>$U=:\{(x,f(x))|x \in \{-0.3, 5\}\}$<br><br>
- **İç Nokta :** Fonksiyon üzerinde tanımlı dış noktalar hariç tüm noktalardır.
<br><br>$İ=:\{(x,f(x))|x \in (-0.3,5)\}$

## **Birinci Türev Bilgisi**

Newton notasyonu:$\ \ f'(x)$
<br>Leibniz notasyonu:$\ \ {{d}\over{dx}}(f(x))$
<br><br>
**Türev$^1$:** Fonksiyonun ***o noktadaki*** eğimi (yönü)
<br>
**Türev$^2$:** Fonksiyonun değerini ***o noktada*** en çok artıran yön
<br>
**Türev$^3$:** Fonksiyonun ***o nokta*** civarında değişken artışına verdiği tepki
<br>
**Türev$^4$:** Fonksiyonun ***o noktada*** artış hızı
<br><br>
***O NOKTA?***
<br><br>
$\ \ {{d}\over{dx}}(f(x))\Bigg|_{x=h}=lim_{h→0}{{f(x+h)-f(x)}\over{h}}$
<br><br>
**Notasyonu okumak:** *x'in f fonksiyonun x'e göre kısmi türevinin h noktasındaki değeri; h 0'a giderken fonksiyonun değer artışının h'ye oranıdır (h başına düşen değer artışıdır).*
<br><br>**O nokta, bu tanımda $(h, f(h))$ noktasıdır (ya da kısaca $h$ noktası).**
____________________
###**Birinci Türev'e Dair Yorumlar ve Kavramlar**
- Birinci türev **pozitif** ise o nokta komşuluğunda fonksiyonun değeri, ilgili argüman değeri artışıyla artar.
 - Fonksiyonun yönü yukarı doğrudur.
<br><br>
- Birinci türev **negatif** ise o nokta komşuluğunda fonksiyonun değeri, ilgili argüman değeri artışıyla azalır.
 - Fonksiyonun yönü aşağı doğrudur.
<br><br>
- Birinci türev **sıfır** ise o nokta komşuluğunda fonksiyonun değeri, ilgili argüman değeri artışıyla değişmez (tepkisizdir).
 - $f'(x)=0$ ve $(x, f(x))$ bir iç nokta ise, $(x, f(x))$ bir ***kritik noktadır***.
 - $f'(x)=0$ ve $(x, f(x))$ bir iç nokta ise, $(x, f(x))$ aynı zamanda bir ***ekstremum*** noktasıdır
- **Ekstremumlar**;
 - $f'(x)=0$ olduğu iç noktalar,
 - $f'(x)$'in tanımlı olmadığı iç noktalar
 <br>ya da
 - uç noktalardır.
 <br><br>
 - **Ekstemum** noktaları ya bir ***yerel minima*** veya bir ***yerel maksima***'dır.
   - $f(\vec x^i)$ eğer bir ***yerel minima*** ise $\vec x^i$ komşuluğunda fonksiyonun daha düşük değer alabildiği başka bir $\vec x^j$ noktası yoktur:
   <br><br>$\ \ \ \ f(\vec x^i)\leq f(\vec x^j)\ ,\ \forall j \in  \{j:|\vec x^i-\vec x^j|\approx \vec 0\}$<br><br>
   - $f(\vec x^i)$ eğer bir ***yerel maksima*** ise $\vec x^i$ komşuluğunda fonksiyonun daha büyük değer alabildiği başka bir $\vec x^j$ noktası yoktur:
   <br><br>$\ \ \ \ f(\vec x^i) \geq f(\vec x^j)\ ,\ \forall j \in  \{j:|\vec x^i-\vec x^j|\approx \vec 0\}$
 <br><br>
 - Bir yerel minima, yerel minimalar arasında en düşük değerli olan ise bir ***global minima***'dır.
 <br><br>$\ \ \ \ f(\vec x^i) \leq f(\vec x^j)\ ,\ \forall j$<br><br>
 - Bir yerel maksima, yerel miksimalar arasında en büyük değerli olan ise bir ***global maksima***'dır.
 <br><br>$\ \ \ \ f(\vec x^i) \geq f(\vec x^j)\ ,\ \forall j$<br><br>
 - Birinci türevi 0 yapan bir noktanın **bükümüne (ikinci türev bilgisine)** bakılarak o noktanın bir yerel minima **veya** yerel maksima olduğu incelenebilir.

## **İkinci Türev Bilgisi**
**Türev (Birinci Türev)**, fonksiyonun *o noktadaki* değişim hızıdır.
<br><br>
**İkinci Türev**, fonksiyonun *o noktadaki* ***değişim hızının değişim hızıdır***. İkinci türev, fonksiyonun bükümü (şekli) ile ilgili bilgi verir. Diğer bir ifade ile fonksiyonun değerleri, o değerleri veren noktalarda birinci ve ikinci türev bilgileri biliniyorsa fonksiyonun grafiği çizilebilir.
- "Türev artansa" (ikinci türev sıfır ya da pozitif) fonksiyon ***yukarıya bükümlüdür***,
- İkinci türevin sıfır ya da pozitif olduğu $[ \vec x^i , \vec x^j ]$ aralığında $f(\vec x)$ ***konvekstir***,
<br>$\ \ \ \ {{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i}\geq 0\ ⇒\ f(x⃗),\ \vec x^i$ komşuluğunda konveks
- Konveks bölgelerde **kritik noktalar** birer **yerel minima**'dır.
<br><br>
- "Türev azalansa" (ikinci türev sıfır ya da negatif) fonksiyon ***aşağı bükümlüdür***,
- İkinci türevin sıfır ya da negatif olduğu $[ \vec x^i , \vec x^j ]$ aralığında $f(\vec x)$ ***konkavdır***,
<br>$\ \ \ \ {{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i}\leq 0\ ⇒\ f(x⃗),\ \vec x^i$ komşuluğunda konkav
- Konveks bölgelerde **kritik noktalar** birer **yerel maksima**'dır.
<br><br>
________________________
**NOT:** Kural olarak konveks bölgelerde ***yerel minima*** noktaları bulunabilir. Yerel minima ya da yerel maksima olmayan kritik noklar bulunabilir.Ancak; ***uç noktaları*** içeren konveks bölgelerde uç noktalarda ve komşuluğunda ***yerel maksima*** noktaları da bulunabilir.
________________________
**NOT:** Kural olarak konkav bölgelerde ***yerel maksima*** noktaları bulunabilir. Yerel minima ya da yerel maksima olmayan kritik noklar bulunabilir.Ancak; ***uç noktaları*** içeren konveks bölgelerde uç noktalarda ve komşuluğunda ***yerel minima*** noktaları da bulunabilir.

##$\ \ \ \ $**Birinci Türev 0 $\Bigg({{d}\over{dx}}(f(\vec x))\Bigg|_{\vec x^i}=0\Bigg)$:**
<br>Bir noktada ($\vec x^i$) fonksiyonun birinci türevi sıfır ise;
- $\vec x^i$ noktası $f(\vec x)$ için bir **durgunluk noktasıdır**. $\vec x^i$ komşuluğunda fonksiyon ne artan ne de azalandır.
- Birinci türevin sıfır olduğu bir $\vec x^i$ noktası için $f(\vec x)$'in ikinci türevi;
 - Pozitif,
 - Sıfır,
 <br>ya da 
 - Negatif<br>
 olabilir.
- **İkinci Türev Pozitif $\Bigg({{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i} > 0\Bigg)$:**
  <br> $f(\vec x)$, $\vec x^i$ komşuluğunda konvekstir ve $\vec x^i$ bir yerel minimadır.
- **İkinci Türev Sıfır $\Bigg({{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i}  = 0 \Bigg)$:**
  <br> $f(\vec x)$ için $\vec x^i$ noktası bir yerel minima, bir yerel maksima olabilir ve de bir [**pervaz**](https://www.desmos.com/calculator/lvvhoovdby)'dır. Bu noktada $f(\vec x)$ hem konveks hem de konkavdır.

- **İkinci Türev Negatif $\Bigg({{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i} < 0\Bigg)$:**
  <br> $f(\vec x)$, $\vec x^i$ komşuluğunda konkavdır ve $\vec x^i$ bir yerel maksimadır.
  <br>
__________________
**SPOILER İÇERİR:** Çok değişkenli fonksiyonlarda bir $\vec x^i$ noktasında $f(\vec x)$, bir $x_{k}$ değişkeni için konveks ve bir diğer $x_{k'}$ değişkeni için ise konkav olabilir. Bu durumda $\vec x^i$ bir **eyer noktasıdır**.
__________________
**NOT:** Tanımı gereği tek değişkenli fonksiyonlar için bir eyer noktası tanımlanamaz.
__________________
##$\ \ \ \ $**Birinci Türev 0'dan farklı $\Bigg({{d}\over{dx}}(f(\vec x))\Bigg|_{\vec x^i}\neq0\Bigg)$:**
<br>Bir noktada ($\vec x^i$) fonksiyonun birinci türevi pozitif ise $f(\vec x$, $\vec x^i$ komşuluğunda **artandır** (yukarı yönlüdür).
- Birinci türevin pozitif olduğu bir $\vec x^i$ noktası için $f(\vec x)$'in ikinci türevi;
 - Pozitif,
 - Sıfır,
 <br>ya da 
 - Negatif<br>

 olabilir.
- **İkinci Türev Pozitif $\Bigg({{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i} > 0\Bigg)$:**
  <br> $f(\vec x)$, $\vec x^i$ komşuluğunda konvekstir.
  <br> $f(\vec x)$'in, $\vec x^i$ komşuluğunda bir kritik noktası yoktur.
- **İkinci Türev Sıfır $\Bigg({{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i}= 0\Bigg)$:**
  <br> $\vec x^i$ noktası $f(\vec x)$ için bir **büküm noktası**'dır.
  <br> $f(\vec x)$'in, $\vec x^i$ komşuluğunda bir kritik noktası yoktur.

- **İkinci Türev Negatif $\Bigg({{d^2}\over{dx^2}}(f(\vec x))\Bigg|_{\vec x^i} < 0\Bigg)$:**
  <br> $f(\vec x)$, $\vec x^i$ komşuluğunda konkavdır.
  <br> $f(\vec x)$'in, $\vec x^i$ komşuluğunda bir kritik noktası yoktur.
__________________
**NOT:** "Dönüm Noktası" kavramını duyabilirsiniz. Dönem noktası yerel minima olan ya da yerel maksima olan kritik noktalar için kullanılan bir kavramdır. **Dönüm noktaları, büküm noktası ile farklı kavramlardır** ve problem çözümü konusunda "ayırt edici" bilgi sunmadığı için bu kavrama, anlatımda yer verilmeyecektir.
__________________
**NOT:** Pervaz noktaları ve büküm noktaları fonksiyonu o noktanın sağında ve solunda farklı işarete sahip ikinci türev değerleri olan bölgelere ayırırlar.
- ***Maksimizasyon*** yaparken;
 - ***konkav bölgelerde***,
 - ***türevi olmayan noktalarda***<br>(o nokta; noktanın sağ ve sol komşuluğunda ikinci türev pozitifse bir yerel maksimadır),
 <br>ve
 - ***dış nokta içeren konveks bölgelerde*** arama yapılır.
<br><br>
- ***Minimizasyon*** yaparken;
 - ***konveks bölgelerde***,
 - ***fonksiyonun sürekli olduğu ancak; türevlenebilir olmayan noktalarda***<br>(o nokta; noktanın sağ ve sol komşuluğunda ikinci  türev negatifse bir yerel minimadır),
 <br>ve
 - ***dış nokta içeren konkav bölgelerde*** arama yapılır.
