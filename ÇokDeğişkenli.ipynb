##**Çok Değişkenli Fonksiyonlar**

Tek değişkenli fonksiyonları incelerken argüman kümesinde tek bir değişken olan ve tek bir görüntü (fonksiyon değeri) elde eden matematiksel ifadeler ele alınmaktadır.<br><br>

Çok değişkenli fonksiyon kavramı ise bağımlı değişkenlerin (görüntü değerleri, $\vec y$) veya bağımsız değişkenlerin (argüman değerleri, $\vec x$) çoklu olması durumunu ifade etmektedir.<br>
Burada vektör gösterimi, çok sayıda değişkenin ve bu değişkenlerin aldığı değerlerin alt alta listelendiğine dair derli toplu bir sunum için tercih edilmektedir:<br><br>

m bağımsız değişkenli ve tek görüntülü bir doğrusal fonksiyon için;<br>
$f(\vec x)=f\left( \begin{bmatrix} x_1\\\vdots\\x_m \end{bmatrix} \right)=f(x_1,\dots,x_m)=\beta_0 + \vec \beta \cdot \vec x=\beta_0 + \beta_1x_1+\dots\beta_mx_m$<br><br>

**NOT:** Önümüzdeki bir kaç ders için vektörleri alt alta değerlerin yazıldığı birer ***liste*** ve matrisleri de transpoze (devrik, yan çevirilmiş) vektörlerin satırlarda yer aldığı bir ***tablo*** olarak görmeniz önerilir.<br><br>

Yukarıda sunulan fonksiyonu görüntü değişkeni olarak $y$ kullanılarak yeniden yazılacak olursa;<br><br>

$y=\beta_0 + \vec \beta \cdot \vec x$<br><br>

biçiminde yazılabilir. Eğer çoklu görüntüden bahsediliyorsa ilgilenilen fonksiyon sonucu vektör olan bir matematiksel ifadedir. Ve bu vektörün her bir bileşeni ($y_1,\dots,y_n$) için birer fonksiyon olduğunu ve bu fonksiyonların ***eşanlı olarak*** aynı bağımsız değişken değerlerine göre değerlendirmek durumunda olduğumuzu düşünmek gerekir.<br><br>

Çok görüntülü fonksiyonlardan şimdilik bu kadar bahsetmek ve zihninizde bu cümlelerin demlenme sürecinde zaman geçirmesini sağlamak daha sağlıklı olacaktır.

###**Tek Görüntülü Çok Değişkenli Fonksiyonlar**

Ders başından bu yana çok defa verdiğimiz "işten ayrılma davranışı fonksiyonu"nu yeniden ancak; bu defa yeni bir perspektiften ziyaret edelim. Önceden fonksiyonun yalnızca tek bir argümana tepkilerini inceliyor ve bu argümanın değerini tasarlayarak süreci yönetmeye çalışıyorduk. Şimdi ise personelimizin işten ayrılışının tek bir argümanla açıklanabilecek kadar basit olmadığını hesaba katıyor ve bu miyopluğumuza karşı bir gözlük kullanıyoruz.<br><br>

Tek bir argümandan ötesini görebilip hesaba katmaya başladığımızda bu çok sayıdaki argümanı eşanlı olarak tasarlamak ve etkinliği (efficiency) yüksek sonuçlar elde etmeye başlayacağız.<br><br>

####**Optimizasyon Yöntemlerine Yeni Bir Değerlendirme**

Şimdiye kadar fonksiyonlarda "*daha mutlu*" olduğumuz bir çözüme ulaşmak için 3 yöntemle tanıştık:
 - Grafik Yöntem
 - Türev Testi
 - Gradyan Arama Algoritması

**Grafik yöntem**, fonksiyonu argüman sayısı kadar boyutu olan bir uzayda görselleştirmek ve bu görselde ekstremum noktalarını karşılaştırmaya dayanmaktadır. Argüman sayısı arttığında, görselleştirilmesi gereken uzay çok boyutlu hale geldiği için 2 değişkenden sonra grafik yöntemin kullanılabilir olmadığnı söyleyebiliriz.<br><br>

**Türev testi**, fonksiyonun birinci türevini 0'a eşitleyen argüman kompozisyonları için ikinci türevin pozitif ya da negatif olmasına göre bulunan kritik noktaların sırasıyla yerel minimum ya da yerel maksimum olarak tanılanmasıdır.
 - Ekstemum noktaları; uç oktalar, türevlenemeyen noktalar ve kritik noktalardır. Türev testi uç noktalara ve türevlenemeyen noktalara dair bir inceleme içermediği için eksiktir.

**Gradyan arama algoritması**, fonksiyon değerini en yakın yerel optimaya doğru adım adım ilerletmek üzere argüman kompozisyonunu sabit bir kurala göre güncellemektir.
 - Gradyan arama algoritmaları;
   - durgunluk noktalarına,
   - türevlenemeyen noktalara,
   - adım büyüklüğü seçimine,
   - başlangıç çözümü seçimine göre kırılganlık içeren bir yöntemler ailesidir.

**Hiçbir yöntem mükemmel değildir.**<br>
Çok değişkenli fonksiyonlarda temel olarak grafik yöntemi kullanamayacağımızı söyleyebiliriz. Türev testi değişken sayısı arttığında maliyeti artarak yükselen bir tercihtir (bu cümleyi daha sonra ayrıntılı olarak açıklayacağız). Gradyan arama algoritmaları ise kırılganlıklar içeren bir yaklaşımdır.<br><br>

###**Çok Değişkenli Fonksiyonlarda Türev Kavramları:<br>Kısmi Türev, Gradyan Vektörü ($\nabla$) ve Yönlü Türev**

Bir hatırlatmayla başlayalım. Leibniz gösterminde bir $f(\vec x)$ fonksiyonunun $x_j$ değişkenine göre kısmi türevi:<br><br>

${d \over dx_j}\left( f(\vec x) \right)$<br><br>

Madem çok sayıda değişkenimiz var. Öyleyse her bir değişkenimize göre (değişken sayısınca) kısmi türev hesaplayabiliriz.<br>
İlk değişkenden başlayarak son değişkene kadar tüm kısmi türevlerin alt alta listelenmesi gradyan vektörüdür:<br><br>

$\nabla \left( f(\vec x) \right) = \begin{bmatrix} 
{d \over dx_1}\left( f(\vec x) \right)\\
\vdots\\
{d \over dx_m}\left( f(\vec x) \right)\\
\end{bmatrix}$<br><br>

Çok değişkenli fonksiyonların bir bağımlı değişkene göre birinci türev bilgisini gradyan vektörü üzerinden inceliyoruz. Burada "*Nabla* ($\nabla$)" gradyan vektörünün sembolik gösterimidir ve $\nabla \left( f(\vec x) \right)$, " *del f* " olarak okunur.<br><br>

Daha önce limit üzerinden kısmi türevini tek değişkenli bir *f* fonksiyonu için $\vec x^0$ noktasında<br><br>

${d \over dx}\left( f(\vec x) \right)\Big|_{\vec x^0} = \lim_{h→0}{f(\vec x^0 + h)-f(\vec x^0)\over h}$<br><br>

biçiminde tanımlamıştık. Çok *m* bağımsız değişkeni olan bir çok değişkenli fonksiyonun j. değişkeni cinsinden kısmi türevini limit üzerinden yazmak için öncelikle bir vektör tanımlayalım:<br><br>

$I^j=\begin{bmatrix} 0\\\vdots\\1\\\vdots\\0 \end{bmatrix}$<br><br>

$I^j$, m bileşeni olan ve j. bileşeni hariç tüm değerleri 0 olan bir vektördür ($m$x$m$ birim matrisin j. sütunudur).<br>
Bu durumda;<br><br>

${d \over dx}\left( f(\vec x) \right)\Big|_{\vec x^0} = \lim_{h→0}{f(\vec x^0 + h\cdot I^j)-f(\vec x^0)\over h}$<br><br>

$f(\vec x)$'in $\ \vec x^0$'deki $\ x_j$'ye göre kısmi türevidir.<br><br>

- **Yönlü Türev** <br><br>

Yönlü türev ("*directional derivatives*"), oldukça önemli ve burada kapsamı oldukça daraltılarak sunulmakta olan bir başlıktır.<br><br>

Yönlü türev, basitçe, bir yön vektörü ($\vec v$) doğrultusunda bağımsız değişkenlerin değer artırımına verilen fonksiyon değeri artışının incelenmesidir.<br><br>

Çok değişkenli fonksiyonlarda kısmi türevin limit üzerinden tanımında kullandığımız $I^j$ vektörü bir yön vektörüdür. $I^j$ yönü, sözel olarak, "yalnızca j. değişkenin arttığı yön" anlamına gelmektedir.<br><br>

*İşletme Matematiği* dersinin misyonunu "yönetmeyi arzu ettiğimiz olguya dair veri toplamak, veriyi açıklayan bir fonksiyonu uyarlamak ve bu fonksiyonun ***kontrolümüz altındaki argümanlarını*** manipüle ederek fonksiyon değerini arzu edilen değerlere taşımayı öğretmek" biçiminde tanıtmıştık.<br><br>

Bu cümleden;
- fonksiyonun tüm argümanlarını kontrol edemeyebileceğimizi,
- fonksiyonun yalnızca seçtiğimiz argümanlarını kontrol etmeyi tercih edebileceğimizi

çıkarımlamalıyız.<br><br>

Öyleyse $I^j$ vektöründe kullanılan mantıktan yola çıkarak; kontrol edemediğimiz veya kontrol etmeyi tercih etmediğimiz değişkenlere karşılık gelen satırlarda *0*, diğer tüm satırlarda "*sıfırdan farklı*" değerlerin olduğu bir $\vec v$ vektörü tanımlayabiliriz:<br><br>

$\nabla_{\vec v}\left(f(\vec x)\right) = \vec v \odot \nabla \left(f(\vec x)\right)$<br><br>

- **Tartışma**<br>

## **Çok Değişkenli Fonksiyonlarda Türev Kavramları (devam):<br>Hessian Matrisi**

Tek görüntülü-çok değişkenli fonksiyonlarda birinci türev bilgisinin gradyan vektörü ($\nabla$) olduğunu belirtmiştik.<br><br>

Öyleyse;
- İkinci türev bilgisi nedir? 
- Nasıl gösterilir? 
- Bir çok değişkenli fonksiyonun konveks ya da konkav olduğu nasıl anlaşılır?<br><br>

Tek değişkenli fonksiyonlarda ikinci türev ardışık olarak fonksiyonun önce kendisinin sonra da birinci türev fonksiyonunun bağımsız değişkene göre kısmi türevi hesaplanarak elde ediliyordu:<br><br>

${d^2 \over dxdx}\left( f(\vec x) \right)={d \over dx}\Big({d \over dx}\left( f(\vec x) \right)\Big)$<br><br>

$m$ değişkenli bir çok değişkenli fonksiyon için birinci türev $m$ ayrı değişkene göre bulunabilir ve bulunan "birinci türev fonksiyonunun" da $m$ ayrı değişkene göre kısmi türevi hesaplanabilir. Yani ikinci türev $m$x$m$ farklı biçimde hesaplanabilir.<br><br>

Önce j. değişkene sonra da k. değişkene göre kısmi türev alınarak elde edilen ikinci türev üzerinden örnek gösterim verilecek olursa:<br><br>

${d^2 \over dx_jdx_k}\left( f(\vec x) \right)={d \over dx_k}\Big({d \over dx_j}\left( f(\vec x) \right)\Big)$<br><br>

Bu $m$x$m$ alternatif sırasıyla önce 1. değişkene göre alınan birinci türev üzerinden 1. değişkene göre ikinci türev, $\dots$ , m. değişkene göre ikinci türev ilk satırda ve son satırda da önce m. değişkene göre alınan birinci türev üzerinden 1. değişkene göre ikinci türev, $\dots$ , m. değişkene göre ikinci türev bilgilerinin yer aldığı bir matris (tablo) biçiminde organize edilebilir:<br><br>

$H_{f(\vec x)} = \begin{bmatrix}
{d^2 \over dx_1dx_1}\left( f(\vec x) \right)&\dots&{d^2 \over dx_1dx_m}\left( f(\vec x) \right)\\
\vdots&\ddots&\vdots\\
{d^2 \over dx_mdx_1}\left( f(\vec x) \right)&\dots&{d^2 \over dx_mdx_m}\left( f(\vec x) \right)
\end{bmatrix}$<br><br>

İkinci türev bilgisinin sunulduğu bu matrisin adı "*Hessian* ($H$)" matrisidir.<br><br>

Hessian matrisinin "pozitif" mi yoksa "negatif" mi olduğuna nasıl bakılır?<br>Şimdilik sadece Hessian'a göre fonksiyonun bir noktada konveks olup olmadığına bakalım.<br><br>
- - -
**Ör.** $f(\vec x)=x^2y$ olsun bu fonksiyonun ikinci türevlerini (1,2) noktasında değerlendiriniz.<br><br>

$\nabla f(\vec x) =\begin{bmatrix}{df\over dx}\\{df\over dy}\end{bmatrix}= \begin{bmatrix}2xy\\x^2\end{bmatrix}$<br><br>
$H_f=\begin{bmatrix}{d^2f \over dxdx}&{d^2f \over dxdy}\\
{d^2f \over dydx}&{d^2f \over dydy}\\\end{bmatrix}=\begin{bmatrix}2y&2x\\2x&0\end{bmatrix}$<br><br>
$H_f(1,2)=\begin{bmatrix}4&2\\2&0\end{bmatrix}$
- - -
**Ör.** Verilen bir A matirisi $\begin{bmatrix}
4&-1\\-1&8
\end{bmatrix}$ biçimindedir. Bu matris bir ikinci türev bilgisi ise ilgili fonksiyonun belirlenen noktada konveks olup olmadığını bulunuz.<br><br>

 - $\Delta_1= \begin{bmatrix}4\end{bmatrix}$, $|\Delta_1| = 4$<br>
 - $\Delta_2= \begin{bmatrix}4&-1\\-1&8\end{bmatrix}$, $|\Delta_1| = 34$<br><br>
A matrisinin temem minörleri için determinant pozitiftir. Diğer bir ifadeyle birinci köşegen boyunca ilk elemanı içeren tüm kare matrisler için determinant pozitiftir. Bu durumda A "*pozitif tanımlıdır*" (positive definite) ve ilgili fonksiyon belirlenen noktada konvekstir.
- - -
**Önemli Not:**<br>Tüm maksimizasyon problemleri için $f(\vec x)$'in en büyük değerini aramak; $g(\vec x)=-f(\vec x)$'in en küçük değerini aramakla denktir.
- Böylece fonksiyonların kritik noktalarda konveks olup olmadığını kontrol edebilmek tüm problemlerin bir yerel optimasını bulmak için yeterlidir.

<br>Hessian matrisini bulmak için $m$ değişkenli bir fonksiyonda $m^2$ işlem yaptık. Ve sonra da $m$ ayrı matrisin determinantını hesapladık...<br>
Çok değişkenli fonksiyonlarda ikinci türev bilgisi üzerinden çalışmak (Newton yaklaşımı) yüksek maliyetlidir.<br><br>

Çok değişkenli fonksiyonlarda özellikle de bağımsız değişken sayısı arttığında türev testi kullanmak zorlu bir tercihtir.

##**Çok Değişkenli Fonksiyonlarda Türev Kavramları (devam):<br>Jacobian Matrisi**

Jacobian matrisi vektör değerli fonksiyonlar için birinci türev bilgisidir. Vektör değerli fonksiyonlar aynı argüman değerleri ile eşanlı olarak hesaplanan çok sayıda fonksiyon olarak düşünülebilir:<br><br>

$\begin{bmatrix}f_1(\vec x)\\\vdots\\f_n(\vec x)\end{bmatrix} =
\begin{bmatrix}
[x_1^2x_2]\\\vdots\\
[x_1+x_2]
\end{bmatrix}$<br><br>

Bu durumda birinci türevler $\nabla_{f_1}=\begin{bmatrix}2x_1x_2\\x_1^2\end{bmatrix}$ ve $\nabla_{f_n}=\begin{bmatrix}1\\1\end{bmatrix}$ olarak bulunabilir.<br><br>

Jacobian matrisi n fonksiyon için herbir satırda ilgili fonksiyona ait gradyan vektörünün transpozunun yer aldığı n satırlı bir matristir.<br><br>

$J(x_1,x_2) = \begin{bmatrix}
{\nabla_{f_1}}^T\\
\vdots \\
{\nabla_{f_n}}^T
\end{bmatrix}=
\begin{bmatrix}
\begin{bmatrix} 2x_1x_2 & x_1^2 \end{bmatrix}\\
\vdots \\
\begin{bmatrix} 1 & 1 \end{bmatrix}
\end{bmatrix}$<br><br>

- **Tartışma**<br>Gradyan arama algoritmasını yenide ziyaret edelim:
 - Çok sayıda değişkeni $\alpha$ ile ölçeklendirilmiş gradyan vektörü doğrultusunda güncellemek:<br>$\vec x^{\ k+1}=\vec x^{\ k} + \nabla(\vec x^{\ k}) \cdot \alpha$<br><br>
 - Çok sayıda değişkeni $\vec \alpha$ ile filtrelenmiş ve ölçeklendirilmiş gradyan vektörü doğrultusunda güncellemek:<br>$\vec x^{\ k+1}=\vec x^{\ k} + \nabla(\vec x^{\ k}) \odot \vec \alpha$<br><br>
 - Çok sayıda değişkeni $\alpha$ ile ölçeklendirilmiş Jacobian matrisi doğrultusunda güncellemek:$\vec x^{\ k+1}=\vec x^{\ k} + J(\vec x^{\ k}) \cdot \alpha$<br><br>
 - Çok sayıda değişkeni $\vec \alpha$ ile filtrelenmiş ve ölçeklendirilmiş Jacobian matrisi doğrultusunda güncellemek:$\vec x^{\ k+1}=\vec x^{\ k} + J(\vec x^{\ k}) \odot \vec \alpha$<br><br>
 - Çok sayıda değişkeni $A_\alpha$ ile filtrelenmiş ve ölçeklendirilmiş Jacobian matrisi doğrultusunda güncellemek:<br>$\vec x^{\ k+1}=\vec x^{\ k} + J(\vec x^{\ k}) \odot A_\alpha$<br><br>
